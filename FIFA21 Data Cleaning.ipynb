### 3) Contract Column

##### This column contained data in 3 different formats such as '2004 ~ 2021', 'Jun 30, 2021 On Loan' and 'Free'.

```python
fifa['Contract'].unique()
```

##### Extract the information from the Contract column and assign it into three new columns: Contract Start, Contract End, and Contract Period (years).

```python
# define a func to extract info from Contract column into 3 variables (to be assigned to 3 new columns later)

def extract_data(contract):

    if contract == 'Free' or 'On Loan' in contract:
        start_date = np.nan
        end_date = np.nan
        contract_period = 0
    else:
        start_date, end_date = contract.split(' ~ ')
        contract_period = int(end_date) - int(start_date)
    return start_date, end_date, contract_period

# define new columns
new_cols = ['Contract Start', 'Contract End', 'Contract Period (years)']
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_data(x)))

# create columns and insert data respectively

for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i, column=new_cols[i], value=new_data[i])
```

```python
fifa['Contract Period (years)'] = fifa['Contract Period (years)'].astype(int)
```

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract Period (years)']].sample(10)
```

##### Categorize each contract status into a new column named "Contract Status."

```python
def status(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'

contract_status = fifa['Contract'].apply(status)

# add contract status column

fifa.insert(fifa.columns.get_loc('Contract Period (years)')+1, 'Contract Status', contract_status)
```

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract Period (years)', 'Contract Status']].sample(10)
```

### 4) Standardize columns with currency €


#### a) Value Column

##### There are two data formats to handle:

##### - Values with a '€' prefix and 'M' suffix (e.g., €103.5M).
##### - Values with a '€' prefix and 'K' suffix (e.g., €9K).

##### After removing the prefix and suffix, the values are converted to float and multiplied by 1,000 for 'K' or 1,000,000 for 'M', then converted to integers.

```python
fifa['Value'].unique()
```

```python
def convert_currency(value):
        if 'K' in value and '€' in value:
            
            euroK = value.replace('€', '').replace('K', '')
            return int(float(euroK)*1000)
        
        else:
            
            euroM = value.replace('€', '').replace('M', '')
            return int(float(euroM)*1000000)

         
fifa['Value'] = fifa['Value'].apply(convert_currency)


```

```python
fifa['Value'].unique()
```

```python
fifa = fifa.rename(columns={'Value':'Value(€)'})
fifa.head(1)
```

### 5) Height Column

#### Convert feet and inches to cm.

```python
fifa['Height'].unique()
```

```python
def convert_height(height):
    if 'cm' in height:
        return int(height.strip('cm'))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches * 2.54)

fifa['Height'] = fifa['Height'].apply(convert_height)
```

```python
fifa['Height'].unique()
```

```python
fifa = fifa.rename(columns={'Height':'Height(cm)'})
```

### 6) Weight Column

#### Convert pounds to kg.

```python
fifa['Weight'].unique()
```

```python
def convert_weight(weight):
    if "kg" in weight:
        return int(weight.strip("kg"))
    else:
        pounds = int(weight.strip('lbs'))
        return round(pounds / 2.205)

# apply func to Weight column

fifa['Weight']= fifa['Weight'].apply(convert_weight)
```

#### b) Hits Column

##### This column is the number of times the player has been searched for in the FIFA Ddatabase.
##### It is possible that there were players who have been not searched, which result in blank records.
##### Those blank records will be filled up with 0 value and change the data type to interger.

```python
fifa[ 'Hits']
```

```python
fifa['Hits'].unique()
```

##### Converting value with 'K' suffix multiple it by 1000 and change the data to integers.



```python
def convert_hits(hits):

    # Hits column consists of null value

    # first, checking on string value with 'K'
    if isinstance(hits, str) and "K" in hits:
        hits_K = float(hits.strip('K')) * 1000
        return int(hits_K) 

    # next, checking on null value and change to 0
    elif pd.isna(hits):
        return 0

    # remaining changed the string to 
    else:
        return int(float(hits))
    
   

# Apply the function to the Hits column
fifa['Hits'] = fifa['Hits'].apply(convert_hits)


```
